\# Using SELECT with LIMIT in PostgreSQL

  

\`LIMIT\` restricts the maximum number of rows returned.  

Examples:

  

\- \`LIMIT 10\` → return at most 10 rows  

\- \`LIMIT 1\` → return a single row  

\- \`LIMIT 0\` → useful for testing query plans  

\- \`LIMIT ... OFFSET ...\` → pagination

  

In this project:

  

\- \`/devices\` API uses \`LIMIT\` for controlled paging.

\- \`/api/v1/latlong\` uses \`LIMIT 1\` to return the nearest location.

```sql
SELECT id, device_id, location, short_address FROM locations;
```

```sql
SELECT version();
```

```sql
SELECT id, asset_id, serial_number, created_at
FROM devices
ORDER BY id
LIMIT 10;
```

```sql
SELECT short_address, long_address
FROM locations
ORDER BY location <-> point(77.5946, 12.9716)
LIMIT 1;
```
